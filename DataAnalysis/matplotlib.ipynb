```python
import matplotlib.pyplot as plt
```

```python
x =[1,2,3,4,5]
y=[5,7,8,3,13]

plt.plot(x,y)
plt.xlabel('Y axis')
plt.show()
```

```python
x =[1,2,3,4,5]
y=[5,7,8,3,13]


plt.plot(x,y, color='red', linestyle='--', marker='o', linewidth=3, markersize=9)
plt.grid(True)
```

```python
## Multiple Plots
## Sample plots

x = [1,2,3,4,5,]
y1= [1,50,9,16,25]
y2= [1,4,8,15,45]

plt.figure(figsize=(9,5))
plt.subplot(2,2,1)
plt.plot(x, y1, color='green')
plt.title('Plot 1')

plt.subplot(2,2,2)
plt.plot(y2, x, color='red')
plt.title('Plot 1')

plt.subplot(2,2,3)
plt.plot(y2, x, color='blue')
plt.title('Plot 1')

plt.subplot(2,2,4)
plt.plot(y2, y1, color='blue')
plt.title('Plot 1')
```

```python
### Bar Plot

categories = ['a','b','c', 'd']
values = [10, 15, 7, 12]
plt.bar(categories, values, color='red')
              
```

```python
## Histogram - Used to represent distribution of data

data = [1,2,2,3,3,3,4,5,5,6,5,6,6]
plt.hist(data, bins=5)
```

```python
## Create a Scatter plot

x = [1,2,3,4,5]
y = [1,4,5,6,9]

plt.scatter(x,y)
```

```python
## Pie Chart

labels = ['A','B','C']
sizes = [30,40,50]
colors=['gold', 'yellowgreen', 'lightcoral']
explode=(0.2,0,0) ## Move out of ist slice

plt.pie(sizes,explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow = True)
```
